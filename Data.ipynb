```python
import datetime
import numpy as np
import pandas as pd
import requests
```

```python
api_url = 'https://ftx.com/api'
api = "/markets"
url = api_url + api
markets = requests.get(url).json()
data = markets['result']
df = pd.DataFrame(data)
df = df.set_index('name')
df
```

```python
start_time = datetime.datetime(2020, 10, 25).timestamp()
end_time = datetime.datetime(2022, 8, 25).timestamp()

api_url = 'https://ftx.com/api'

market_name = 'ETH/USD'

resolution = (60*60*24)

path = f'/markets/{market_name}/candles?resolution={resolution}&start_time={start_time}&end_time={end_time}'
url = api_url + path
res = requests.get(url).json()
df = pd.DataFrame(res['result'])
print(df) 
```
